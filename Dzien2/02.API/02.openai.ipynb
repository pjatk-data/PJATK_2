W tym ćwiczeniu użyjemy biblioteki OpenAI dla Pythona, aby wykonać te same operacje, co w pierwszym ćwiczeniu.

Najpierw używamy instrukcji `import`, aby nasza aplikacja wiedziała, że w kodzie będziemy korzystać z biblioteki OpenAI.

```python
%pip install openai
```

```python
import openai
```

Następnie zaczniemy wczytywać wartości z pliku `.env`.

```python
import os
from dotenv import load_dotenv

# Wczytanie zmiennych środowiskowych
if load_dotenv():
    print("Znaleziono bazowy punkt końcowy Azure OpenAI API: " + os.getenv("AZURE_OPENAI_ENDPOINT"))
else: 
    print("Nie znaleziono bazowego punktu końcowego Azure OpenAI API. Czy skonfigurowano plik .env?")
```

Utworzymy nowy obiekt `AzureOpenAI` i przekażemy do niego klucz API, wersję oraz adres endpointu, którego chcemy użyć.

```python
from openai import OpenAI

client = OpenAI(
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    base_url=os.getenv("AZURE_OPENAI_ENDPOINT")
)
```

## Wysyłanie promptu do Azure OpenAI z użyciem biblioteki OpenAI

Skoro zdefiniowaliśmy już instancję Azure OpenAI, spróbujmy wykonać Chat Completion. Wywołamy metodę `chat.completions.create()`. Zwróć uwagę, że jako wartość `model` przekazujemy w rzeczywistości identyfikator naszego wdrożenia Azure OpenAI (`deployment`). Przekażemy również prompt, którego chcemy użyć, jako `content` parametru `messages`.

```python
response = client.chat.completions.create(
    model = os.getenv("AZURE_OPENAI_COMPLETION_MODEL"),
    messages = [{"role" : "user", "content" : "Co oznacza liczba 42?"}],
)

print(response)
```

Powyższa odpowiedź powinna mieć strukturę JSON podobną do tej, którą otrzymaliśmy przy bezpośrednim wywołaniu API w poprzednim ćwiczeniu. Zawiera ona szczegóły modelu, wygenerowaną odpowiedź oraz zużycie tokenów.

Możemy wykorzystać ustrukturyzowaną odpowiedź zwróconą przez API, aby wyodrębnić sam wygenerowany tekst odpowiedzi.

```python
print(response.choices[0].message.content)
```

Teraz wypiszemy tekst odpowiedzi w taki sam sposób, jak zrobiliśmy to w poprzednim ćwiczeniu.

```python
import json
response_json = json.loads(response.model_dump_json())
print(json.dumps(response_json, indent=4))
```

## Podsumowanie

Łatwiej też dodać do wywołania informacje takie jak prompt, ponieważ możemy przekazać je bezpośrednio jako argumenty metod biblioteki OpenAI, zamiast dołączać obiekty JSON do treści żądania.
