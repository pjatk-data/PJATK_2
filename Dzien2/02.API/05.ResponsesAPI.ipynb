# Responses API
Responses API to nowe stanowe API od Azure OpenAI. Łączy najlepsze możliwości z API chat completions i assistants w jednym spójnym doświadczeniu. Responses API dodaje też obsługę nowego modelu computer-use-preview, który zasila funkcję Computer use.

## Dokumentacja API
- [Dokumentacja Responses API](https://learn.microsoft.com/en-us/azure/foundry/openai/latest#create-response)

```python
import os
import logging
from openai import OpenAI
from dotenv import load_dotenv

# Konfiguracja logowania do konsoli
logger = logging.getLogger()
logging.basicConfig(level=logging.INFO)

if load_dotenv():
    url = os.getenv('AZURE_OPENAI_ENDPOINT')
    key = os.getenv("AZURE_OPENAI_API_KEY")
    model = os.getenv("AZURE_OPENAI_COMPLETION_MODEL")

    logger.info(f"Używany punkt końcowy: {url}")
    logger.info(f"Używany klucz: {key[:4]}...{key[-4:]}")
    logger.info(f"Używany model: {model}")

else:
    logger.error("Nie znaleziono pliku .env")
    exit(1)

client = OpenAI(
    api_key=key,
    base_url=url,
)

response = client.responses.create(
    model=model,  # Zastąp nazwą wdrożonego modelu
    input="This is a test.",
)

print(response.model_dump_json(indent=2))
```

# Pobieranie odpowiedzi
Aby pobrać odpowiedź z poprzedniego wywołania Responses API.

```python
response_id=response.id
response_retrived = client.responses.retrieve(response_id)
print(response_retrived.model_dump_json(indent=2))
```

# Łączenie odpowiedzi
Możesz łączyć odpowiedzi, przekazując response.id z poprzedniej odpowiedzi do parametru previous_response_id.

```python
response = client.responses.create(
    model=model,
    input="Zdefiniuj i krótko wyjaśnij pojęcie katastrofalnego zapominania.",
)


second_response = client.responses.create(
    model=model,
    previous_response_id=response.id,
    input="Wyjaśnij to na poziomie zrozumiałym dla studenta pierwszego roku studiów.",
)

print("Pierwsza odpowiedź:")
print(response.output_text)

print("Druga odpowiedź:")
print(second_response.output_text)
```

```python
response_id=second_response.id
response_retrived = client.responses.retrieve(response_id)
print(response_retrived.model_dump_json(indent=2))
```
