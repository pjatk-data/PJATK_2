# Modele rozumowania Azure OpenAI

Modele rozumowania Azure OpenAI są zaprojektowane do wykonywania zadań wymagających wnioskowania i rozwiązywania problemów z większym skupieniem i możliwościami. Modele te poświęcają więcej czasu na przetwarzanie i zrozumienie prośby użytkownika, dzięki czemu są wyjątkowo skuteczne w takich obszarach jak nauka, programowanie i matematyka w porównaniu z wcześniejszymi wersjami.

Najważniejsze możliwości modeli rozumowania:

- Generowanie złożonego kodu: Potrafią tworzyć algorytmy i obsługiwać zaawansowane zadania programistyczne wspierające deweloperów.
- Zaawansowane rozwiązywanie problemów: Idealne do kompleksowych burz mózgów i rozwiązywania wieloaspektowych wyzwań.
- Złożone porównywanie dokumentów: Świetnie nadają się do analizowania umów, akt spraw czy dokumentów prawnych w celu wykrywania subtelnych różnic.
- Przestrzeganie instrukcji i zarządzanie przepływem pracy: Szczególnie skuteczne w zarządzaniu przepływami pracy wymagającymi krótszego kontekstu.




```python
import os
import logging
from openai import OpenAI
from dotenv import load_dotenv

logger = logging.getLogger()
logging.basicConfig(level=logging.INFO)

load_dotenv()

url = os.getenv('AZURE_OPENAI_ENDPOINT')
key = os.getenv("AZURE_OPENAI_API_KEY")
model = os.getenv("AZURE_OPENAI_COMPLETION_MODEL")

client = OpenAI(
    api_key=key,
    base_url=url
)

response = client.chat.completions.create(
    model=model, # zastąp nazwą wdrożonego modelu dla swojego wdrożenia o1.
    messages=[
        {"role": "user", "content": "Jakie kroki powinienem rozważyć, pisząc swoje pierwsze API w Pythonie?"},
    ],
    max_completion_tokens = 5000

)

print(response.model_dump_json(indent=2))
```

# Nakład rozumowania
Modele rozumowania mają reasoning_tokens jako część completion_tokens_details w odpowiedzi modelu. Są to ukryte tokeny, które nie są zwracane jako część treści odpowiedzi wiadomości, ale są używane przez model do wygenerowania końcowej odpowiedzi na Twoją prośbę. reasoning_effort można ustawić na low, medium lub high dla wszystkich modeli rozumowania z wyjątkiem o1-mini. Modele rozumowania GPT-5 obsługują nową wartość reasoning_effort o nazwie minimal. Im wyższy poziom nakładu, tym dłużej model będzie przetwarzał żądanie, co zwykle skutkuje większą liczbą reasoning_tokens.

```python
response = client.chat.completions.create(
    model=model, 
    messages=[
        {"role": "developer","content": "Jesteś pomocnym asystentem."}, 
        {"role": "user", "content": "Jakie kroki powinienem rozważyć, pisząc swoje pierwsze API w Pythonie?"},
    ],
    max_completion_tokens = 5000,
    reasoning_effort = "medium" # niski, średni lub wysoki
)

print(response.model_dump_json(indent=2))
```

# Podsumowanie rozumowania
Podczas korzystania z najnowszych modeli rozumowania z Responses API możesz użyć parametru reasoning summary, aby otrzymywać podsumowania łańcucha rozumowania modelu.

```python
response = client.responses.create(
    input="Opowiedz mi o ciekawym przypadku degeneracji tekstu neuronowego", 
    model=model, # zastąp nazwą wdrożonego modelu
    reasoning={
        "effort": "medium",
        "summary": "auto" # automatycznie, zwięźle lub szczegółowo, serie gpt-5 nie obsługują opcji concise 
    },
    text={
        "verbosity": "low" # nowe w modelach GPT-5
    }
)

print(response.model_dump_json(indent=2))
```
