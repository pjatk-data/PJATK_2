# 01 - Praca bezpośrednio z interfejsem API Azure OpenAI

W tym ćwiczeniu wykonamy kilka prostych wywołań do interfejsu API Azure OpenAI.
- Pierwsze wywołanie pozwoli sprawdzić, które wdrożenia modeli są dostępne w Azure OpenAI API.
- Drugie wywołanie prześle prompt do Azure OpenAI API.

Dzięki temu sprawdzimy również, czy wszystko jest poprawnie skonfigurowane i gotowe do pracy w kolejnych ćwiczeniach.

## Konfiguracja

Najpierw musimy pobrać wartości z pliku `.env`, których użyjemy do wywołań Azure OpenAI API.

```python
%pip install python-dotenv
%pip install requests
```

```python
import json
import requests
import os
from dotenv import load_dotenv


# Wczytanie zmiennych środowiskowych
if load_dotenv():
    print("Znaleziono bazowy punkt końcowy Azure OpenAI API: " + os.getenv("AZURE_OPENAI_ENDPOINT"))
else: 
    print("Nie znaleziono bazowego punktu końcowego Azure OpenAI API. Czy skonfigurowano plik .env?")
    
API_KEY = os.getenv("AZURE_OPENAI_API_KEY")
RESOURCE_ENDPOINT = os.getenv("AZURE_OPENAI_ENDPOINT")
```

## Wysyłanie promptu do Azure OpenAI z użyciem API

Teraz wywołamy Azure OpenAI API z promptem. Do tego potrzebujemy `id` wdrożenia Azure OpenAI, które zawiera nasz model do uzupełniania. Powinno ono być już zapisane w pliku `.env`.

```python
DEPLOYMENT_ID = os.getenv("AZURE_OPENAI_COMPLETION_MODEL")
```

Zbudujemy adres URL wywołania, aby zobaczyć, jak wygląda.

```python
url = RESOURCE_ENDPOINT +  "chat/completions"

print(url)
```

Następnie wywołamy Azure OpenAI API z użyciem powyższego adresu URL. Klucz API przekażemy w nagłówku HTTP. W treści żądania wyślemy również dane w formacie JSON, które zawierają *prompt* użyty do uzyskania odpowiedzi od modelu OpenAI. W tym przypadku naszym promptem jest "Dawno, dawno temu", co powinno spowodować, że model rozwinie go, generując opowieść.

```python
r = requests.post(url, 
                  headers={"api-key": API_KEY}, 
                  json={
      "model": DEPLOYMENT_ID,
      "messages": [
      {
        "role": "developer",
        "content": "Jesteś pomocnym asystentem."
      },
      {
        "role": "user",
        "content": "Wyjaśnij, czym jest gorzka lekcja."
      }]
    })

print(json.dumps(r.json(), indent=2))
```



Poniżej znajduje się kilka informacji o bardziej interesujących elementach danych zawartych w tej odpowiedzi.

Pole | Opis
--- | ---
`model` | Model użyty do wygenerowania odpowiedzi na prompt
`content` | To odpowiedź wygenerowana przez model OpenAI. Zauważ, że przykład odpowiedzi powyżej został skrócony, aby był łatwiejszy do odczytania
`finish_reason` | Powód, dla którego model zakończył generowanie odpowiedzi. W tym przypadku `stop` oznacza, że model uznał odpowiedź za w pełni ukończoną i nie napotkał ograniczeń
`completion_tokens` | Liczba tokenów użytych do wygenerowania odpowiedzi
`prompt_tokens` | Liczba tokenów zużytych przez prompt
`total_tokens` | Łączna liczba tokenów zużytych przez żądanie (`prompt_tokens` + `completion_tokens`)
