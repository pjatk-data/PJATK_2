# Podobieństwo cosinusowe
Podobieństwo cosinusowe to miara używana do określania, jak podobne są dwa wektory, niezależnie od ich długości. Jest powszechnie stosowane w kontekście embeddingów, czyli reprezentacji punktów danych w przestrzeni o wysokiej liczbie wymiarów, takich jak embeddingi słów w przetwarzaniu języka naturalnego.

### Najważniejsze informacje o podobieństwie cosinusowym:

1. **Definicja**:
   Podobieństwo cosinusowe oblicza cosinus kąta między dwoma wektorami. Wzór jest następujący:
   $$
   \text{podobieństwo cosinusowe} = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}
   $$
   gdzie \( \mathbf{A} \cdot \mathbf{B} \) oznacza iloczyn skalarny wektorów \( \mathbf{A} \) i \( \mathbf{B} \), a \( \|\mathbf{A}\| \) i \( \|\mathbf{B}\| \) oznaczają długości (lub normy) wektorów.

2. **Zakres**:
   Wartość podobieństwa cosinusowego mieści się w przedziale od -1 do 1:
   - **1** oznacza, że wektory są identyczne.
   - **0** oznacza, że wektory są prostopadłe (brak podobieństwa).
   - **-1** oznacza, że wektory są dokładnie przeciwne.

3. **Zastosowania**:
   - **Analiza tekstu**: Służy do porównywania podobieństwa między dokumentami lub zdaniami.
   - **Systemy rekomendacyjne**: Pomaga znajdować podobne elementy lub użytkowników na podstawie ich embeddingów.
   - **Rozpoznawanie obrazów**: Służy do porównywania wektorów cech obrazów.

4. **Zalety**:
   - **Niezależność od skali**: Podobieństwo cosinusowe nie zależy od długości wektorów, dzięki czemu jest przydatne przy porównywaniu danych znormalizowanych.
   - **Wydajność**: Jest obliczeniowo efektywne dla danych wielowymiarowych.

5. **Przykład**:
   Załóżmy, że mamy dwa embeddingi słów \( \mathbf{A} \) i \( \mathbf{B} \):
   $$
   \mathbf{A} = [1, 2, 3], \quad \mathbf{B} = [4, 5, 6]
   $$
   Podobieństwo cosinusowe oblicza się następująco:
   $$
   \text{podobieństwo cosinusowe} = \frac{(1 \cdot 4 + 2 \cdot 5 + 3 \cdot 6)}{\sqrt{1^2 + 2^2 + 3^2} \cdot \sqrt{4^2 + 5^2 + 6^2}} = \frac{32}{\sqrt{14} \cdot \sqrt{77}} \approx 0.974
   $$

Podobieństwo cosinusowe to potężne narzędzie do mierzenia podobieństwa w różnych zastosowaniach, szczególnie przy pracy z danymi o wysokiej liczbie wymiarów, takimi jak embeddingi.

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

# Obliczanie embeddingów dla modelu ada

```python

text = 'the quick brown fox jumped over the lazy dog'
model = 'text-embedding-ada-002'
try:
    text_embedding_ada=client.embeddings.create(input = [text], model=model, dimensions=1536).data[0].embedding
except Exception as e:
    text_embedding_ada=client.embeddings.create(input = [text], model=model,).data[0].embedding
```

```python
print(text_embedding_ada[:10])
print(len(text_embedding_ada))
```

# Porównanie słów z użyciem embeddingów i podobieństwa cosinusowego

```python
# porównaj kilka słów
automobile_embedding_ada    = client.embeddings.create(input = 'automobile', model=model).data[0].embedding
vehicle_embedding_ada       = client.embeddings.create(input = 'vehicle', model=model).data[0].embedding
dinosaur_embedding_ada      = client.embeddings.create(input = 'dinosaur', model=model).data[0].embedding
stick_embedding_ada         = client.embeddings.create(input = 'stick', model=model).data[0].embedding
huskey_embedding_ada        = client.embeddings.create(input = 'huskey', model=model).data[0].embedding
brown_embedding_ada         = client.embeddings.create(input = 'brown fox', model=model).data[0].embedding

# porównując podobieństwo cosinusowe, samochód względem samochodu powinien dać 1.0, czyli dokładnie to samo, natomiast samochód względem dinozaura powinien dać wartość między 0 a 1, czyli nie to samo
print(cosine_similarity(automobile_embedding_ada, automobile_embedding_ada))
print(cosine_similarity(automobile_embedding_ada, vehicle_embedding_ada))
print(cosine_similarity(automobile_embedding_ada, dinosaur_embedding_ada))
print(cosine_similarity(automobile_embedding_ada, stick_embedding_ada))
print(cosine_similarity(automobile_embedding_ada, huskey_embedding_ada))
```
